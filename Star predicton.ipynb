```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

# step2: Load Dataset

```python
ss=pd.read_csv("C:\\Users\\Dell\\OneDrive\\Desktop\\jup\\star project logestic\\stars.csv")
ss
```

# plot graph

```python
import seaborn as sns
sns.pairplot(ss)
```

# Step 3: Seperate input and output in different variables

```python
ll = LabelEncoder()
ss["Star color"]=ll.fit_transform(ss["Star color"])
ss["Spectral Class"]=ll.fit_transform(ss["Spectral Class"])
ss
```

```python
ss["Star type"].unique()
```

```python
x=ss.drop(columns=["Star type"])
x
```

```python
y=ss.iloc[:,-3]
y
```

# Step 4:Split input output data in training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size= 0.2,random_state= 42)
```

# Step5:Import Algorithm

```python
from sklearn. linear_model import LogisticRegression
pen=LogisticRegression()
```

# Step6:Model Train(.fit()) and Prediction(.predict())Â¶

```python
pen.fit(x_train,y_train)
```

```python
y_pred=pen.predict(x_test)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
